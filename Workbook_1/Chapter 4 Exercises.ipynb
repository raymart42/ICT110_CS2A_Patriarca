#### Exercise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

```python
def compute_pay(hours, rate):
  if hours > 40:
    regular_pay = 40 * rate
    overtime_hours = hours - 40
    overtime_pay = overtime_hours * rate * 1.5
    total_pay = regular_pay + overtime_pay
  else:
    total_pay = hours * rate

  return total_pay


hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

pay = compute_pay(hours, rate)
print("Pay:", pay)
```

## CHALLENGE

#### Challenge

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

```python
def compute_grade(score):

  if score < 0 or score > 100:
    return "Invalid score"

  elif score >= 97:
    return "1.0"
  elif score >= 93:
    return "1.25"
  elif score >= 89:
    return "1.5"
  elif score >= 85:
    return "1.75"
  elif score >= 81:
    return "2.0"
  elif score >= 77:
    return "2.25"
  elif score >= 74:
    return "2.5"
  elif score >= 0:
    return "3.0"

score = int(input("Enter Score: "))

grade = compute_grade(score)
print("Grade:", grade)
```
